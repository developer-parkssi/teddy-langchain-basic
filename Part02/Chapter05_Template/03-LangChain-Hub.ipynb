## 10 LangChain Hub에서 프롬프트 공유하기

```python
from langchain import hub

# 가장 최신 버전의 프롬프트를 가져옵니다.
prompt = hub.pull("rlm/rag-prompt")
print(prompt)
```

- 프롬프트를 가져올 때 버전 해시를 가져올 수 있음
- LangChain Hub의 Commits 탭을 클릭하면 버전 지정

```python
prompt = hub.pull("rlm/rag-prompt:50442af1")
print(prompt)
```

- 자신의 프롬프트를 LangChain Hub에 업로드하기

```python
from langchain.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_template(
    "주어진 내용을 바탕으로 다음 문장을 요약하세요. 답변은 반드시 한글로 작성하세요\n\nCONTEXT: {context}\n\nSUMMARY:"
)
prompt
```

- push 메서드 안에 자신의 LangChain ID를 입력하고 프롬프트 저장소 이름 지정

```python
from dotenv import load_dotenv

# 환경변수 로드
load_dotenv()

from langchain import hub

# Hub에 프롬프트 업로드
# 참고: username 없이 프롬프트 이름만 사용하면 자동으로 본인 계정에 업로드됩니다
hub.push("simple-summary-korean", prompt)
```
